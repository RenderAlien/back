```python
import requests
```

### Регистрация

```python
user_signup_json = {
    'first_name': 'first_name',
    'second_name': 'second_name',
    'email': 'email',
    'password': 'password',
    'adress': 'adress',
    'is_admin': False
}

r = requests.post('http://localhost:80/api/auth/signup', json=user_signup_json).json()
if 'error' in r:
    print('ERROR', r['error'])
else:
    token, uid = r['token'], r['uid']

admin_signup_json = {
    'first_name': 'admin',
    'second_name': 'admin',
    'email': 'admin',
    'password': 'admin',
    'adress': 'admin',
    'is_admin': True
}

r = requests.post('http://localhost:80/api/auth/signup', json=admin_signup_json).json()
if 'error' in r:
    print('ERROR', r['error'])
else:
    admin_token, admin_uid = r['token'], r['uid']
```

```python
headers = {'Authorization': 'Bearer ' + token}
admin_headers = {'Authorization': 'Bearer ' + admin_token}
```

### Создание категории

```python
json = {
    'name': 'Milks'
}
```

```python
r = requests.post('http://localhost:80/api/catalog/createcategory', json=json, headers=headers).json()
r
```

```python
r = requests.post('http://localhost:80/api/catalog/createcategory', json=json, headers=admin_headers).json()
r
```

```python
category_id = r['category_id']
```

### Создание продукта

```python
json = {
    'name': 'Cheese',
    'desc': 'Good cheese.',
    'price': 351.8,
    'category_id': category_id,
    'quantity': 15
}
```

```python
r = requests.post('http://localhost:80/api/catalog/createproduct', json=json, headers=headers).json()
r
```

```python
r = requests.post('http://localhost:80/api/catalog/createproduct', json=json, headers=admin_headers).json()
r
```

### Выход

```python
r = requests.post('http://localhost:80/api/auth/signout', json={'token': token}, headers=headers).json()
r
```

```python
r = requests.post('http://localhost:80/api/auth/signout', json={'token': admin_token}, headers=admin_headers).json()
r
```
